#### Delete method
##### Due to a lack of resources and time, we were not able to obtain the entire data for the entire timeframe, hence we cannot fill in any of the missing values. To accomodate for this limitation we made a "cleaning" method that deletes all the rows with empty values.

```python
def clean_csv_by_deleting_rows(input_folder, output_folder=None, overwrite_original=False):
    if not os.path.exists(input_folder):
        print(f"Error: Input folder '{input_folder}' does not exist.")
        return

    if not overwrite_original:
        if output_folder is None:
            output_folder = os.path.join(input_folder, 'Cleaned_data')
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)
        print(f"Cleaned files will be saved to: {output_folder}")
    else:
        print("Warning: Original files will be overwritten!")


 

    if not csv_files:
        print(f"No CSV files found in '{input_folder}'.")
        return

    print(f"Found {len(csv_files)} CSV files to process.")

    for i, filename in enumerate(csv_files):
        file_path = os.path.join(input_folder, filename)
        print(f"\nProcessing file {i+1}/{len(csv_files)}: {filename}")

        try:
            # Read the CSV
            df = pd.read_csv(file_path)

            original_rows = len(df)
            df.dropna(inplace=True)
            cleaned_rows = len(df)

            print(f"  Original rows: {original_rows}")
            print(f"  Rows after deleting NaNs: {cleaned_rows}")
            print(f"  Deleted {original_rows - cleaned_rows} rows.")

            # Save the cleaned DataFrame
            if overwrite_original:
                output_path = file_path
            else:
                output_path = os.path.join(output_folder, filename)

            df.to_csv(output_path, index=False) # index=False prevents writing the DataFrame index as a column
            print(f"  Cleaned data saved to: {output_path}")

        except Exception as e:
            print(f"  Error processing {filename}: {e}")

    print("\nAll files processed.")

# --- Configuration ---
# IMPORTANT: Replace 'your_csv_folder_path' with the actual path to your CSVs
INPUT_CSV_FOLDER = 'your_csv_folder_path'

OUTPUT_CSV_FOLDER = None # Not used if overwrite_original is True
OVERWRITE_ORIGINAL_FILES = False # Set to True to overwrite originals

# --- Run the cleaning script ---
clean_csv_by_deleting_rows(
    input_folder=INPUT_CSV_FOLDER,
    output_folder=OUTPUT_CSV_FOLDER,
    overwrite_original=OVERWRITE_ORIGINAL_FILES
)
```

### Fix: Change the price action to price change

```python
nvda_RSIMACD['3 Day price change']= nvda_RSIMACD['Open'].shift(-3)-nvda_RSIMACD['Open']
nvda_RSIMACD['5 Day price change']= nvda_RSIMACD['Open'].shift(-5)-nvda_RSIMACD['Open']
nvda_RSIMACD['7 Day price change']= nvda_RSIMACD['Open'].shift(-7)-nvda_RSIMACD['Open']
nvda_RSIMACD['10 Day price change']= nvda_RSIMACD['Open'].shift(-10)-nvda_RSIMACD['Open']
nvda_RSIMACD['14 Day price change']= nvda_RSIMACD['Open'].shift(-14)-nvda_RSIMACD['Open']
nvda_RSIMACD.tail(14)
```

```python
#getting rid of the NaNs
last_five_columns = nvda_RSIMACD[-5:]
for col_name in last_five_columns:
    nvda_RSIMACD[col_name] = nvda_RSIMACD[col_name].fillna(nvda_RSIMACD['Open'])
columns_to_drop= nvda_RSIMACD.columns[0:5]
nvda_fixed_RSIMACD = nvda_RSIMACD.drop(columns = columns_to_drop)
nvda_fixed_RSIMACD.tail(14)
```

```python
corrRSI = nvda_fixed_RSIMACD.corr()
corrRSI
```

```python
plt.figure(figsize=(10,10))
sb.heatmap(corrRSI, cmap='RdBu_r',vmin = -0.2, vmax=1, annot = True,square = True, annot_kws={'fontsize':6, 'fontweight': 'bold'})
plt.title("Correlation between price and MACD, RSI, and news strategy")
```
